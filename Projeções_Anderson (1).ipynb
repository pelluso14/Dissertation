```python
# Autor: Anderson Luiz Silva Pelluso
# Suporte técnico: Bernardo Lacerda Salgueiro Faria
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel('/content/Cenários_dissertação_EPE.xlsm', sheet_name = 'Potencial energético')
df.head()
```

```python
historico = df[['Unnamed: 3']][2:27]
```

```python
baubau = df[['Unnamed: 3']][26:]
bau_moderado = df[['Unnamed: 7']][26:]
bau_acelerada = df[['Unnamed: 11']][26:]
```

```python
moderada_bau = df[['Unnamed: 15']][26:]
moderada_moderada = df[['Unnamed: 19']][26:]
moderada_acelerada = df[['Unnamed: 23']][26:]
```

```python
acelerado_bau = df[['Unnamed: 27']][26:]
acelerado_moderada = df[['Unnamed: 31']][26:]
acelerado_acelerada = df[['Unnamed: 35']][26:]
```

```python
# Criar o vetor de anos
anos_historico = pd.date_range(start="2000-01-01", end="2024-01-01", freq='YS')  # Histórico anual
anos_pred = pd.date_range(start="2023-12-31", end="2050-01-01", freq='YS')  #

# Plotar os dados
figure = plt.figure(figsize=(10, 5))
plt.plot(anos_historico, historico, marker='o', color='black', label='Histórico')
plt.plot(anos_pred, baubau, marker='o', color='red', label='BAU BAU')
plt.plot(anos_pred, bau_moderado, marker='o', color='blue', label='BAU Moderado')
plt.plot(anos_pred, bau_acelerada, marker='o', color='green', label='BAU Acelerada')
plt.plot(anos_pred, moderada_bau, marker='o', color='red', label='Moderado BAU')
plt.plot(anos_pred, moderada_moderada, marker='o', color='blue', label='Moderado Moderado')
plt.plot(anos_pred, moderada_acelerada, marker='o', color='green', label='Moderado Acelerado')
plt.plot(anos_pred, acelerado_bau, marker='o', color='red', label='Acelerado BAU')
plt.plot(anos_pred, acelerado_moderada, marker='o', color='blue', label='Acelerado Moderado')
plt.plot(anos_pred, acelerado_acelerada, marker='o', color='green', label='Acelerado Acelerado')

# Adicionar título, rótulos e grade
plt.title("Exemplo de gráfico com anos de 2000 a 2050")
plt.xlabel("Ano")
plt.ylabel("Valor")
plt.grid()
plt.legend()  # Mostrar legenda
plt.show()
```

```python
baubau = np.array([float(x) if x is not None else 0.0 for x in baubau.values.flatten()])
bau_moderado = np.array([float(x) if x is not None else 0.0 for x in bau_moderado.values.flatten()])
bau_acelerada = np.array([float(x) if x is not None else 0.0 for x in bau_acelerada.values.flatten()])
moderada_bau = np.array([float(x) if x is not None else 0.0 for x in moderada_bau.values.flatten()])
moderada_moderada = np.array([float(x) if x is not None else 0.0 for x in moderada_moderada.values.flatten()])
moderada_acelerada = np.array([float(x) if x is not None else 0.0 for x in moderada_acelerada.values.flatten()])
acelerado_bau = np.array([float(x) if x is not None else 0.0 for x in acelerado_bau.values.flatten()])
acelerado_moderada = np.array([float(x) if x is not None else 0.0 for x in acelerado_moderada.values.flatten()])
acelerado_acelerada = np.array([float(x) if x is not None else 0.0 for x in acelerado_acelerada.values.flatten()])
```

```python
# Gráfico
plt.figure(figsize=(10, 8))

# Linha histórica
plt.plot(anos_historico, historico, color='black', label="Histórico")

# Preenchimento com cores destacadas e opacidades ajustadas
plt.fill_between(anos_pred, acelerado_moderada, acelerado_bau, color='green', alpha=0.4, edgecolor='darkgreen', label="Moderada - Acelerado")
plt.fill_between(anos_pred, acelerado_moderada, acelerado_acelerada, color='green', alpha=0.6, edgecolor='darkgreen', label="Moderada - Baubau")
plt.fill_between(anos_pred, moderada_bau, moderada_moderada, color='blue', alpha=0.4, edgecolor='darkblue', label="Acelerado - Moderada")
plt.fill_between(anos_pred, moderada_moderada, moderada_acelerada, color='blue', alpha=0.6, edgecolor='darkblue', label="Moderada - Moderado")
plt.fill_between(anos_pred, baubau, bau_moderado, color='red', alpha=0.4, edgecolor='darkred', label="Acelerado - Moderada")
plt.fill_between(anos_pred, bau_moderado, bau_acelerada, color='red', alpha=0.6, edgecolor='darkred', label="Moderada - Acelerado")
# Melhorando o gráfico
#plt.title("Exemplo de gráfico com anos de 2000 a 2050", fontsize=14, fontweight='bold')
#plt.xlabel("Ano", fontsize=12)
#plt.ylabel("Valor", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
#plt.legend(loc="upper left", fontsize=10)
plt.show()
```

```python
# Gráfico
plt.figure(figsize=(12, 8))

# Linha histórica
plt.plot(anos_historico, historico, color='black', linewidth=2, label="Histórico")

# Preenchimento com cores destacadas e opacidades ajustadas
# A área de preenchimento verde (acelerado_moderada vs acelerado_bau)
plt.fill_between(anos_pred, acelerado_moderada, acelerado_bau,
                 linestyle='-', linewidth=2, color='green', alpha=0.6, edgecolor='darkgreen', label=None)

# A área de preenchimento verde mais clara (acelerado_acelerada vs acelerado_moderada)
plt.fill_between(anos_pred, acelerado_moderada, acelerado_acelerada,
                 linestyle='-', linewidth=2, color='green', alpha=0.4, edgecolor='darkgreen', label="Elasticidade = 1.59")

# A área de preenchimento azul (moderada_bau vs moderada_moderada)
plt.fill_between(anos_pred, moderada_bau, moderada_moderada,
                 linestyle='-', color='blue', linewidth=2, alpha=0.6, edgecolor='darkblue', label=None)

# A área de preenchimento azul mais clara (moderada_acelerada vs moderada_moderada)
plt.fill_between(anos_pred, moderada_moderada, moderada_acelerada,
                 linestyle='-', linewidth=2, color='blue', alpha=0.4, edgecolor='darkblue', label="Elasticidade = 1.29")

# A área de preenchimento vermelha (baubau vs bau_moderado)
plt.fill_between(anos_pred, baubau, bau_moderado,
                 linestyle='-', linewidth=2, color='red', alpha=0.6, edgecolor='darkred', label=None)

# A área de preenchimento vermelha mais clara (bau_acelerada vs bau_moderado)
plt.fill_between(anos_pred, bau_moderado, bau_acelerada,
                 linestyle='-', linewidth=2, color='red', alpha=0.4, edgecolor='darkred', label="Elasticidade = 1")

# Melhorando o gráfico com uma grade
plt.grid(True, linestyle='--', alpha=0.6)

# Adicionando a legenda
plt.legend(loc="upper left", fontsize=10, title="Legenda")

# Exibindo o gráfico
plt.show()
```
