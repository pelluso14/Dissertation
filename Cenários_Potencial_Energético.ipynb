```python
# Autor: Anderson Luiz Silva Pelluso
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_excel('Cenários_dissertação_PC.xlsm', sheet_name='Potencial energético')
df.head()
```

```python
historico = df[['Unnamed: 3']][2:27]
```

```python
bau_base = df[['Unnamed: 3']][26:]
bau_moderada = df[['Unnamed: 7']][26:]
bau_acelerada = df[['Unnamed: 11']][26:]
```

```python
retracao_base = df[['Unnamed: 15']][26:]
retracao_moderada = df[['Unnamed: 19']][26:]
retracao_acelerada = df[['Unnamed: 23']][26:]
```

```python
conexoes_base = df[['Unnamed: 27']][26:]
conexoes_moderada = df[['Unnamed: 31']][26:]
conexoes_acelerada = df[['Unnamed: 35']][26:]
```

```python
# Criar o vetor de anos
anos_historico = pd.date_range(start="2000-01-01", end="2024-01-01", freq='YS')  # Histórico anual
anos_pred = pd.date_range(start="2023-12-31", end="2050-01-01", freq='YS')  #

# Plotar os dados
figure = plt.figure(figsize=(10, 5))
plt.plot(anos_historico, historico, marker='', color='black', label='Histórico')
plt.plot(anos_pred, bau_base, marker='x', color='red', label='BAU base')
plt.plot(anos_pred, bau_moderada, marker='o', color='red', label='BAU Moderada')
plt.plot(anos_pred, bau_acelerada, marker='s', color='red', label='BAU Acelerada')
plt.plot(anos_pred, retracao_base, marker='x', color='blue', label='Retração Base')
plt.plot(anos_pred, retracao_moderada, marker='o', color='blue', label='Retração Moderada')
plt.plot(anos_pred, retracao_acelerada, marker='s', color='blue', label='Retração Acelerada')
plt.plot(anos_pred, conexoes_base, marker='x', color='green', label='Conexões Base')
plt.plot(anos_pred, conexoes_moderada, marker='o', color='green', label='Conexões Moderada')
plt.plot(anos_pred, conexoes_acelerada, marker='s', color='green', label='Conexões Acelerada')

# Adicionar título, rótulos e grade
plt.title("Exemplo de gráfico com anos de 2000 a 2050")
plt.xlabel("Ano")
plt.ylabel("Valor")
plt.grid()
plt.legend()  # Mostrar legenda
plt.show()
```

```python
bau_base = np.array([float(x) if x is not None else 0.0 for x in bau_base.values.flatten()])
bau_moderada = np.array([float(x) if x is not None else 0.0 for x in bau_moderada.values.flatten()])
bau_acelerada = np.array([float(x) if x is not None else 0.0 for x in bau_acelerada.values.flatten()])
retracao_base = np.array([float(x) if x is not None else 0.0 for x in retracao_base.values.flatten()])
retracao_moderada = np.array([float(x) if x is not None else 0.0 for x in retracao_moderada.values.flatten()])
retracao_acelerada = np.array([float(x) if x is not None else 0.0 for x in retracao_acelerada.values.flatten()])
conexoes_base = np.array([float(x) if x is not None else 0.0 for x in conexoes_base.values.flatten()])
conexoes_moderada = np.array([float(x) if x is not None else 0.0 for x in conexoes_moderada.values.flatten()])
conexoes_acelerada = np.array([float(x) if x is not None else 0.0 for x in conexoes_acelerada.values.flatten()])
```

```python

sns.set(style="whitegrid")

# Melhorar a visualização dos dados
plt.style.use('seaborn-darkgrid')
# Melhorar a visualização dos dados
plt.style.use('seaborn-darkgrid')


sns.set(style="whitegrid")

# Criar a figura
plt.figure(figsize=(14, 10))

# Linha histórica
sns.lineplot(x=anos_historico, y=historico.values.flatten(), color='black', linewidth=4, label="Histórico")

# Preenchimento com cores destacadas e opacidades ajustadas
plt.fill_between(anos_pred, conexoes_moderada, conexoes_base, color='green', alpha=0.4, label=None)
plt.fill_between(anos_pred, conexoes_moderada, conexoes_acelerada, color='green', alpha=0.4, label=None)
plt.fill_between(anos_pred, retracao_base, retracao_moderada, color='blue', alpha=0.4, label=None)
plt.fill_between(anos_pred, retracao_moderada, retracao_acelerada, color='blue', alpha=0.4, label=None)
plt.fill_between(anos_pred, bau_base, bau_moderada, color='red', alpha=0.4, label=None)
plt.fill_between(anos_pred, bau_moderada, bau_acelerada, color='red', alpha=0.4, label=None)

# Adicionando as curvas com marcadores específicos
sns.lineplot(x=anos_pred, y=conexoes_base, marker='x', linestyle='--', color='green', linewidth=2, label='Conexões Base')
sns.lineplot(x=anos_pred, y=conexoes_moderada, marker='o', linestyle='-', color='green', linewidth=2, label='Conexões Moderada')
sns.lineplot(x=anos_pred, y=conexoes_acelerada, marker='s', linestyle='-.', color='green', linewidth=2, label='Conexões Acelerada')
sns.lineplot(x=anos_pred, y=retracao_base, marker='x', linestyle='--', color='blue', linewidth=2, label='Retração Base')
sns.lineplot(x=anos_pred, y=retracao_moderada, marker='o', linestyle='-', color='blue', linewidth=2, label='Retração Moderada')
sns.lineplot(x=anos_pred, y=retracao_acelerada, marker='s', linestyle='-.', color='blue', linewidth=2, label='Retração Acelerada')
sns.lineplot(x=anos_pred, y=bau_base, marker='x', linestyle='--', color='red', linewidth=2, label='BAU Base')
sns.lineplot(x=anos_pred, y=bau_moderada, marker='o', linestyle='-', color='red', linewidth=2, label='BAU Moderada')
sns.lineplot(x=anos_pred, y=bau_acelerada, marker='s', linestyle='-.', color='red', linewidth=2, label='BAU Acelerada')

# Melhorando o gráfico
plt.title("Cenários de Potencial Energéticos (2000-2050)", fontsize=16, fontweight='bold')
plt.xlabel("Ano", fontsize=14)
plt.ylabel("Valor (MJ)", fontsize=14)
plt.grid(True, linestyle='--', alpha=0.4)

# Adicionando a legenda com elasticidades do PIB
handles, labels = plt.gca().get_legend_handles_labels()
elasticidade_labels = ['Elasticidade_PIB = 1', 'Elasticidade_PIB = 0.52', 'Elasticidade_PIB = 0.71 ']
for color, label in zip(['green', 'blue', 'red'], elasticidade_labels):
    handles.append(plt.Line2D([0], [0], color=color, lw=4))
    labels.append(label)
plt.legend(handles, labels, loc="upper left", fontsize=12, title="Cenários")

plt.xticks(rotation=45)
plt.tight_layout()

# Exibindo o gráfico
plt.show()
```

```python
# Gráfico
plt.figure(figsize=(12, 8))

# Linha histórica
plt.plot(anos_historico, historico, color='black', linewidth=2, label="Histórico")

# Preenchimento com cores destacadas e opacidades ajustadas
# A área de preenchimento verde (acelerado_moderada vs acelerado_bau)
plt.fill_between(anos_pred, conexoes_moderada, conexoes_base,
                 linestyle='-', linewidth=2, color='green', alpha=0.4, edgecolor='darkgreen', label=None)

# A área de preenchimento verde mais clara (acelerado_acelerada vs acelerado_moderada)
plt.fill_between(anos_pred, conexoes_moderada, conexoes_acelerada,
                 linestyle='-', linewidth=2, color='green', alpha=0.4, edgecolor='darkgreen', label="Elasticidade = 1.59")

# A área de preenchimento azul (moderada_bau vs moderada_moderada)
plt.fill_between(anos_pred, retracao_base, retracao_moderada,
                 linestyle='-', color='blue', linewidth=2, alpha=0.6, edgecolor='darkblue', label=None)

# A área de preenchimento azul mais clara (moderada_acelerada vs moderada_moderada)
plt.fill_between(anos_pred, retracao_moderada, retracao_acelerada,
                 linestyle='-', linewidth=2, color='blue', alpha=0.4, edgecolor='darkblue', label="Elasticidade = 1.29")

# A área de preenchimento vermelha (baubau vs bau_moderado)
plt.fill_between(anos_pred, bau_base, bau_moderada,
                 linestyle='-', linewidth=2, color='red', alpha=0.6, edgecolor='darkred', label=None)

# A área de preenchimento vermelha mais clara (bau_acelerada vs bau_moderado)
plt.fill_between(anos_pred, bau_moderada, bau_acelerada,
                 linestyle='-', linewidth=2, color='red', alpha=0.4, edgecolor='darkred', label="Elasticidade = 1")

# Melhorando o gráfico com uma grade
plt.grid(True, linestyle='--', alpha=0.6)

# Adicionando a legenda
plt.legend(loc="upper left", fontsize=10, title="Legenda")

# Exibindo o gráfico
plt.show()
```
